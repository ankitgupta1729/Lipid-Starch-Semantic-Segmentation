```python
import os, sys
from PIL import Image
```

```python
input_path="//SIDCTSTMATSTUD1/RnD_Polymer/ComputerVision_Analytical/Algae TEM Images/Set 2/"
output_path="//SIDCTSTMATSTUD1/RnD_Polymer/ComputerVision_Analytical/Algae TEM Images/Set 2_JPEG/"
```

```python
def tiff_to_jpeg():
    for infile in os.listdir(input_path):
        print("file name: " + infile)
        if infile[-4:] == "tiff":
            outfile = infile[:-4] + "jpeg"
            img = Image.open(input_path+infile)
            print("new filename : " + outfile)
            out = img.convert("RGB")
            out.save(output_path+outfile, "JPEG", quality=90)
        if infile[-3:] == "tif":
            outfile = infile[:-3] + "jpeg"
            img = Image.open(input_path+infile)
            print("new filename : " + outfile)
            out = img.convert("RGB")
            out.save(output_path+outfile, "JPEG", quality=90)
```

```python
tiff_to_jpeg()
```
